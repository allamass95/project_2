```python
def divided_pop(table,party,year):
    table=table.loc[table['leg_control']==f'{party}']
    table["Population"]=table["Population"].astype(float)
    min_value=table["Population"].min()
    table['Population']=table['Population']/min_value
    
    table['Median Household Income']=table['Median Household Income']*table['Population']
    table['Per Capita Income']=table['Per Capita Income']*table['Population']
    table['Poverty Rate']=table['Poverty Rate']*table['Population']
    table['Unemployment Rate']=table['Unemployment Rate']*table['Population']
    table['Median Cost of Rent']=table['Median Cost of Rent']*table['Population']
    table['Black Poverty Rate']=table['Black Poverty Rate']*table['Population']
    table['Hispanic Poverty Rate']=table['Hispanic Poverty Rate']*table['Population']
    table['White Poverty Rate']=table['White Poverty Rate']*table['Population']
    
    avg_hou_income=table['Median Household Income'].sum()/table['Population'].sum()
    avg_per_cap_income=table['Per Capita Income'].sum()/table['Population'].sum()
    avg_poverty_rate=table['Poverty Rate'].sum()/table['Population'].sum()
    avg_unem_rate=table['Unemployment Rate'].sum()/table['Population'].sum()
    avg_rent=table['Median Cost of Rent'].sum()/table['Population'].sum()
    avg_black=table['Black Poverty Rate'].sum()/table['Population'].sum()
    avg_Hispanic=table['Hispanic Poverty Rate'].sum()/table['Population'].sum()
    avg_white=table['White Poverty Rate'].sum()/table['Population'].sum()
    
    
    list_ahi=[]
    list_apc=[]
    list_apr=[]
    list_aur=[]
    list_ar=[]
    list_ab=[]
    list_ah=[]
    list_aw=[]
    
    list_ahi.append(avg_hou_income)
    list_apc.append(avg_per_cap_income)
    list_apr.append(avg_poverty_rate)
    list_aur.append(avg_unem_rate)
    list_ar.append(avg_rent)
    list_ab.append(avg_black)
    list_ah.append(avg_Hispanic)
    list_aw.append(avg_white)
    
    
    
    list_ahi=pd.DataFrame(list_ahi)
    list_apc=pd.DataFrame(list_apc)
    list_apr=pd.DataFrame(list_apr)
    list_aur=pd.DataFrame(list_aur)
    list_ar=pd.DataFrame(list_ar)
    list_ab=pd.DataFrame(list_ab)
    list_ah=pd.DataFrame(list_ah)
    list_aw=pd.DataFrame(list_aw)
    
    list_ahi['Median Household Income']=list_ahi
    list_apc['Per Capita Income']=list_apc
    list_apr['Poverty Rate']=list_apr
    list_aur['Unemployment Rate']=list_aur
    list_ar['Median Cost of Rent']=list_ar
    list_ab['Black Poverty Rate']=list_ab
    list_ah['Hispanic Poverty Rate']=list_ah
    list_aw['White Poverty Rate']=list_aw
    
    
    da_table=pd.concat([list_ahi,list_apc,list_apr,list_aur,list_ar,list_ab,list_ah,list_aw],axis=1)
    da_table['year']=year
    da_table=da_table.drop(0,axis=1)
    return(da_table)
    
```

```python
r_2011=divided_pop(finalized_2011,"Rep",'2011')
r_2012=divided_pop(finalized_2012,"Rep",'2012')
r_2013=divided_pop(finalized_2013,"Rep",'2013')
r_2014=divided_pop(finalized_2014,"Rep",'2014')
r_2015=divided_pop(finalized_2015,"Rep",'2015')
r_2016=divided_pop(finalized_2016,"Rep",'2016')
r_2017=divided_pop(finalized_2017,"Rep",'2017')

final_rep_data=pd.concat([r_2011,r_2012,r_2013,r_2014,r_2015,r_2016,r_2017],axis=0).reset_index(drop=True)
```
