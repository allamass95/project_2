```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
db="postgres:Alevaporo58@localhost:5432/project_2"
engine=create_engine(f'postgresql://{db}')
```

```python
conn=engine.connect()
```

```python
stats_2011=pd.read_sql('select * from year_2011',conn)
stats_2012=pd.read_sql('select * from year_2012',conn)
stats_2013=pd.read_sql('select * from year_2013',conn)
stats_2014=pd.read_sql('select * from year_2014',conn)
stats_2015=pd.read_sql('select * from year_2015',conn)
stats_2016=pd.read_sql('select * from year_2016',conn)
stats_2017=pd.read_sql('select * from year_2017',conn)
```

```python
def filtered_table(table):
    table=table[['Median Household Income',
                 'Per Capita Income',
                 'Poverty Rate',
                 'Unemployment Rate',
                 'Median Cost of Rent',
                 'Black Poverty Rate',
                 'Hispanic Poverty Rate',
                 'White Poverty Rate']]
    
    table=pd.DataFrame(table)
    return(table)
```

```python
table_2011=filtered_table(stats_2011)
table_2012=filtered_table(stats_2012)
table_2013=filtered_table(stats_2013)
table_2014=filtered_table(stats_2014)
table_2015=filtered_table(stats_2015)
table_2016=filtered_table(stats_2016)
table_2017=filtered_table(stats_2017)
```

```python
def index_clean(table):
 
    table=table.drop(table.index[[8,51]])
    table=table.reset_index(drop=True)
    table=pd.DataFrame(table)
    return(table)
```

```python
clean_table_2011=index_clean(table_2011)
clean_table_2012=index_clean(table_2012)
clean_table_2013=index_clean(table_2013)
clean_table_2014=index_clean(table_2014)
clean_table_2015=index_clean(table_2015)
clean_table_2016=index_clean(table_2016)
clean_table_2017=index_clean(table_2017)

```

```python
def add_year(table,year):
    
    table['date']=year
    return(table)
```

```python
new_table_2011=add_year(clean_table_2011,'2011')
new_table_2012=add_year(clean_table_2012,'2012')
new_table_2013=add_year(clean_table_2012,'2013')
new_table_2014=add_year(clean_table_2012,'2014')
new_table_2015=add_year(clean_table_2012,'2015')
new_table_2016=add_year(clean_table_2012,'2016')
new_table_2017=add_year(clean_table_2012,'2017')
```

```python
states=pd.read_sql('select state,leg_control from data_2011',conn)
```

```python
graph_data_2011=pd.concat([states,clean_table_2011],axis=1)
graph_data_2012=pd.concat([states,clean_table_2012],axis=1)
graph_data_2013=pd.concat([states,clean_table_2013],axis=1)
graph_data_2014=pd.concat([states,clean_table_2014],axis=1)
graph_data_2015=pd.concat([states,clean_table_2015],axis=1)
graph_data_2016=pd.concat([states,clean_table_2016],axis=1)
graph_data_2017=pd.concat([states,clean_table_2017],axis=1)
```

```python
def groupby(table):
    table=table.groupby(table['leg_control']).mean()
    return(table)
```

```python
final_2011=groupby(graph_data_2011)
final_2012=groupby(graph_data_2012)
final_2013=groupby(graph_data_2013)
final_2014=groupby(graph_data_2014)
final_2015=groupby(graph_data_2015)
final_2016=groupby(graph_data_2016)
final_2017=groupby(graph_data_2017)
```

```python
final_2011['date']='2011'
final_2012['date']='2012'
final_2013['date']='2013'
final_2014['date']='2014'
final_2015['date']='2015'
final_2016['date']='2016'
final_2017['date']='2017'
```

```python
complete_data=pd.concat([final_2011,final_2012,final_2013,final_2014,final_2015,final_2016,final_2017],axis=0)
complete_data=complete_data.sort_values(['leg_control','date'],ascending=True)
```
